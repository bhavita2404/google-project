```python
# breaking the route line to decrease the total error

import numpy as np 
import matplotlib.pyplot as plt
import gmplot
  
def estimate_coef(x1, y1):
    x = np.array(x1)
    y = np.array(y1)
    
    # number of points 
    n = np.size(x) 
  
    # mean of x and y array
    m_x, m_y = np.mean(x), np.mean(y) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 
  
    # calculating regression coefficients 
    b1 = SS_xy / SS_xx 
    b0 = m_y - b1*m_x 
  
    return(b0, b1)  


def plot_regression_route(lat,lon,gmap1,pcolor,name):

    # estimating coefficients 
    b = estimate_coef(lat,lon) 
    #print(b[0],b[1])
    
    # plotting regression line 
    y_pred = b[0] + b[1]*np.array(lat)
    
    neg = 0
    x1,y1,x2,y2=[],[],[],[]
    for i in range(len(lon)):
        if lon[i] - y_pred[i] < 0:
            neg+=1
            x1.append(lat[i])
            y1.append(lon[i])
        else:
            if neg >= 4:
                x1.append(lat[i])
                y1.append(lon[i])
                b1 = estimate_coef(x1,y1)
                y_pred1 = b1[0] + ( b1[1]* np.array(x1) )
                gmap1.plot(x1,y_pred1,pcolor,edge_width=2.5)
                x1,y1=[],[]
                neg = 0
            elif neg > 0:
                x2 = x2 + x1
                y2 = y2 + y1
                x1,y1 = [],[]
                neg = 0
                
            x2.append(lat[i])
            y2.append(lon[i])
        print(name[i],(lon[i]-y_pred[i]))
    
    if neg >= 4 :
        x2.append(x1[0])
        y2.append(y1[0])
        b1 = estimate_coef(x1,y1)
        y_pred1 = b1[0] + ( b1[1]* np.array(x1) )
        gmap1.plot(x1,y_pred1,pcolor,edge_width=2.5)
    elif neg > 0 and neg < 4:
        x2 = x2 + x1
        y2 = y2 + y1
        
    b2 = estimate_coef(x2,y2)
    y_pred2 = b2[0] + ( b2[1]*np.array(x2))
    #gmap1.apikey = "AIzaSyAbr98I2YDjSiJR5Sv81zkfrB5uNAhCuMU"
    gmap1.plot(x2,y_pred2,pcolor,edge_width=2.5)
    

def main():

    gmap1 = gmplot.GoogleMapPlotter(18.77778,73.31833, 10 ) 
    
    file3 = open('C:\\Users\\S\\OneDrive\\Desktop\\local_train_route_coord.txt',"r")
    route_lat,route_lon,name = [],[],[]
    count = 0
    for line in file3:
        if count == 0:
            count+=1
            continue;  
        color = line.split('\t')[12]
        if color == "blue":
            try:
                route_lat.append(float(line.split('\t')[10]))
                route_lon.append(float(line.split('\t')[11][:-1]))
                name.append((line.split('\t')[9]))
            except:
                continue #only for spaces
        count+=1  
    gmap1.scatter(route_lat,route_lon,'white',size = 100, marker = False)
    plot_regression_route(route_lat,route_lon,gmap1,"blue",name)
    gmap1.draw( "C:\\Users\\S\\OneDrive\\Desktop\\LR(improved).html")
    
if __name__ == "__main__": 
    main()
```
