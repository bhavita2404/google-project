```python
import numpy as np 
import matplotlib.pyplot as plt
import gmplot
  
def estimate_coef(x, y): 
    # number of observations/points 
    n = np.size(x) 
  
    # mean of x and y vector 
    m_x, m_y = np.mean(x), np.mean(y) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 
  
    # calculating regression coefficients 
    b_1 = SS_xy / SS_xx 
    b_0 = m_y - b_1*m_x 
  
    return(b_0, b_1)  

def plot_regreesion_route(lat,lon,gmap1,pcolor):
    
    x = np.array(lat)
    # estimating coefficients 
    b = estimate_coef(x, np.array(lon)) 
    print(b[0],b[1])
    
    # plotting regression line 
    y_pred = b[0] + b[1]*x
    
    gmap1.scatter(lat,lon,'white',size = 100, marker = False)
    #gmap1.apikey = "AIzaSyAbr98I2YDjSiJR5Sv81zkfrB5uNAhCuMU"
    gmap1.plot(x,y_pred,pcolor,edge_width=2.5)
    
def main():
    
    file = open("C:\\Users\\S\\Desktop\\lat_long.txt", "r")
    lat,lon = [],[]
    count = 0
    for line in file:
        if count == 0:
            count+=1
            continue;
        lat.append(float(line.split('|')[2]))
        lon.append(float(line.split('|')[3][:-1]))

    gmap1 = gmplot.GoogleMapPlotter(18.77778,73.31833, 10 ) 
    
    file3 = open('C:\\Users\\S\\Desktop\\local_train_route_coord.txt',"r")
    route_lat,route_lon = [],[]
    count = 0
    route = 1
    color = ""
    for line in file3:
        if count == 0:
            count+=1
            continue;  
        x = int(line.split('\t')[13])
        if count == 1:
            route = int(x)
            color = line.split('\t')[12]
            try:
                route_lat.append(float(line.split('\t')[10]))
                route_lon.append(float(line.split('\t')[11][:-1]))
            except:
                continue #only for spaces
        else:
            if x == route:
                try:
                    route_lat.append(float(line.split('\t')[10]))
                    route_lon.append(float(line.split('\t')[11][:-1]))
                except:
                    continue #only for spaces
            else:
                route = int(x)
                pcolor = color 
                color = line.split('\t')[12]
                print(pcolor)
                
                plot_regreesion_route(route_lat,route_lon,gmap1,pcolor)
                route_lat,route_lon = [],[]
        count+=1  

    plot_regreesion_route(route_lat,route_lon,gmap1,pcolor)
    gmap1.draw( "C:\\Users\\S\\Desktop\\map14.html")
    
if __name__ == "__main__": 
    main() 
```
