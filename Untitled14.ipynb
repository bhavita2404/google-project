```python
import shapely
import math
from shapely.geometry import Point
from shapely.geometry import LineString
from math import radians, cos, sin, asin, sqrt

class Graph:
    
    def __init__(self):
        self.adjacency_list = dict()
        
    def add_edge(self,src,dest):
        if src not in self.adjacency_list:
            self.adjacency_list[src] = []
        self.adjacency_list[src].append(dest)
        
        if dest not in self.adjacency_list:
            self.adjacency_list[dest] = []
        self.adjacency_list[dest].append(src)
        
    def print_graph(self):
        for x in self.adjacency_list:
            print (x , "->" , self.adjacency_list[x])
            
    def find_degree(self,val):
        if val in self.adjacency_list:
            l = set(self.adjacency_list[val])
            return len(l)
        else:
            return 0
        

def dist_bw2points(lat1,lon1,lat2,lon2):
    # to calculate Distance Between Two Points on Earth

    # converts from degrees to radians.
    lon1 = radians(lon1)
    lon2 = radians(lon2)
    lat1 = radians(lat1)
    lat2 = radians(lat2)

    # Haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles

    return (c*r)        
        
def create_offset(points,gmap1):
    lat1,lon1,lat2,lon2 = points[0][0],points[0][1],points[1][0],points[1][1]
    dist = dist_bw2points(lat1,lon1,lat2,lon2)
    offsetPixels = 0.1
    if dist == 0:
        dist = 1
    lat1p = lat1 + offsetPixels * (lon2-lon1) / dist
    lat2p = lat2 + offsetPixels * (lon2-lon1) / dist
    lon1p = lon1 + offsetPixels * (lat1-lat2) / dist
    lon2p = lon2 + offsetPixels * (lat1-lat2) / dist
    
    latitudes = [lat1p,lat2p]
    longitudes = [lon1p,lon2p]
    gmap1.plot(latitudes, longitudes , color = 'red', edge_width = 2)

def error_check(route_lat,route_lon,points,name,train_no,csv_file,gmap1) :
    latitudes = [points[0][0],points[1][0]]
    longitudes = [points[0][1],points[1][1]]
    gmap1.plot(latitudes, longitudes , color = 'yellow', edge_width = 2)
    
    create_offset(points,gmap1)
    
    csv_file.write(str(train_no) + "," + name[0] + "," + '0' + '\n')
    for i in range(1,len(route_lat)-1): 
        point = Point(route_lat[i], route_lon[i])
        dist = LineString([points[0],points[1]]).project(point)
        l = list(LineString([points[0],points[1]]).interpolate(dist).coords)
        err_dist = dist_bw2points(route_lat[i],route_lon[i],l[0][0],l[0][1])
        csv_file.write(str(train_no) + "," + name[i] + "," + str(err_dist) + '\n')
    csv_file.write(str(train_no) + "," + name[len(route_lon)-1] + "," + '0' + '\n')
    
if __name__ == "__main__":
    
    g = Graph()
    
    csv_file = open("C:\\Users\\S\\OneDrive\\Desktop\\local_mumbai_err_dist.txt", "a")
    csv_file.write("train_no" + "," + "location_name" + "," + "err_dist" + '\n')
    
    data = pd.read_csv('C:\\Users\\S\\OneDrive\\Desktop\\lat_long.txt', sep='|')
    vertices = data['station_code'].tolist()
    
    data['lat'] = data['lat'].astype(float)
    data['lng'] = data['lng'].astype(float)
    
    latitudes = data['lat'].tolist()
    longitudes = data['lng'].tolist()
    
    
    file = open('C:\\Users\\S\\OneDrive\\Desktop\\mumbai_seperated_train_route_coord.txt',"r")
    count = 0
    next(file)
    train_no = 1
    name, prev_n = "",""
    for line in file:  
        x = line.split('\t')[0]
        if x == "green \n" or x == "blue \n" or x =="violet \n" or x =="red \n":
            continue;
        x = int(x)
        prev_n = name
        name = line.split('\t')[1]
        if count == 0:
            count+=1
            train_no = int(x)
        else:
            if x == train_no:
                g.add_edge(name,prev_n)
            else:
                train_no = int(x)
        count+=1
       
    junc_endpoint = {}
    for i in range(len(vertices)):
        x = g.find_degree(vertices[i])
        if x > 2 or x == 1 :
            junc_endpoint[vertices[i]] = i
    
    gmap1 = gmplot.GoogleMapPlotter(18.77778,73.31833, 10 ) 
    
    route_lat, route_lon,location_name = [],[],[]
    junc_to_junc_stations_list = []
    file.seek(0,0)
    points = []
    count,name = 0,""
    train_no = 1
    for line in file:  
        x = line.split('\t')[0]
        if x == "green \n" or x == "blue \n" or x =="violet \n" or x =="red \n":
            continue;
        x = int(x)
        prev_name = name
        name = line.split('\t')[1]
        print(name)
        lat,lon = float(line.split('\t')[10]),float(line.split('\t')[11][:-1])
        if count == 0:
            count+=1
            train_no = int(x)
            print(train_no)
            if name in junc_endpoint:
                points.append((latitudes[junc_endpoint[name]] , longitudes[junc_endpoint[name]]))
                route_lat.append(lat)
                route_lon.append(lon) 
                location_name.append(name)
        else:
            if x == train_no:
                if name in junc_endpoint:
                    route_lat.append(lat)
                    route_lon.append(lon) 
                    location_name.append(name)
                    points.append((latitudes[junc_endpoint[name]] , longitudes[junc_endpoint[name]]))
                    if len(points) == 2 :
                        error_check(route_lat, route_lon,points,location_name,train_no,csv_file,gmap1)
                        points,route_lat,route_lon,location_name = [],[],[],[]
                        points.append((latitudes[junc_endpoint[name]] , longitudes[junc_endpoint[name]]))
                else:
                    route_lat.append(lat)
                    route_lon.append(lon) 
                    location_name.append(name)
                    
            else:
                train_no = int(x)
                if len(points) == 2:
                    error_check(route_lat,route_lon,points,location_name,train_no,csv_file,gmap1)
                    
                points,route_lat, route_lon,location_name = [],[],[],[]
                
                if name in junc_endpoint:
                    points.append((latitudes[junc_endpoint[name]] , longitudes[junc_endpoint[name]])) 
                    route_lat.append(lat)
                    route_lon.append(lon) 
                    location_name.append(name)
               
        count+=1 
        
    gmap1.draw( "C:\\Users\\S\\OneDrive\\Desktop\\offset1.html") 
    csv_file.close()
    
```
